```
monk_gln = prototype(verbose=1);
monk_gln.Prototype("Task", "resnet50");

# Insert data and set params in default mode
monk_gln.Default(dataset_path="/content/car_data/train", 
            model_name="resnet50", 
            freeze_base_network=True,
            num_epochs=10);
```

```
monk_gln.Train();
```

```
monk_eval=prototype(verbose=1)
monk_eval.Prototype("Task", "resnet50", eval_infer=True);
```
